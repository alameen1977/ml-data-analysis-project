Build a classification model without hyperparameter tuning using an algorithm of your choice

````python
#fetch dataset and create train/test split
#We obtain the credit-default dataset and separate predictors from the label.
#We remove an ID column if present, ensure the label is integer, and split stratified to keep class balance.

import pandas as pd
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split

# Retrieve dataset (skip this fetch if you already loaded it earlier)
ds = fetch_ucirepo(name='Default of Credit Card Clients')

# Features and label
features_df = ds.data.features.copy()
label_series = ds.data.targets.iloc[:, 0].copy().astype(int)   # ensure 0/1 integer labels

# Remove non-informative ID if it exists
if 'ID' in features_df.columns:
    features_df = features_df.drop(columns=['ID'])

# Simple check for missing values (stratify requires no NaNs in label)
if features_df.isna().any().any() or label_series.isna().any():
    print("Warning: missing values found — consider imputing before training")

# Stratified split: keep the same class proportions in both sets
train_X, test_X, train_y, test_y = train_test_split(
    features_df, label_series, test_size=0.30, stratify=label_series, random_state=123
)

print(f"Total rows: {len(features_df)}")
print(f"Training rows: {len(train_X)}, Test rows: {len(test_X)}")
print("Positive class proportion (whole dataset):", label_series.mean().round(4))
````

````python
#I loaded the Credit Card Default dataset
#I used a stratified 70/30 split so both train and test keep the same proportion of defaulters.
# I ensured the label is numeric and removed the ID column because it does not help prediction.
````

````python
#scale numeric features and train a Decision Tree baseline
# - Decision trees do not require scaling, but we keep a scaler step commented in case you try other models.
# - We train a simple tree as a quick baseline.

from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

# Note: DecisionTreeClassifier doesn't need scaling, but scaler shown for comparison with other models
scaler = StandardScaler()
# scaled_train_X = scaler.fit_transform(train_X)   # uncomment if you use a model that needs scaling
# scaled_test_X  = scaler.transform(test_X)

# Train a small Decision Tree baseline (no hyperparameter tuning)
dt = DecisionTreeClassifier(random_state=123)
dt.fit(train_X, train_y)   # pass raw features; tree handles numeric variables directly

print("Trained Decision Tree. Number of features used:", train_X.shape[1])
````

````python
#I trained a Decision Tree as a baseline because it is fast and easy to interpret.
# I did not scale inputs because trees are insensitive to feature scaling.
# Next I will evaluate performance on the test set.”
````

````python
# evaluate basic metrics on the test set
#We compute accuracy, precision, recall and show the confusion matrix counts.
#For imbalanced problems, precision and recall are more informative than accuracy.

from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Predictions
y_pred_test = dt.predict(test_X)

# Metrics
acc = accuracy_score(test_y, y_pred_test)
prec = precision_score(test_y, y_pred_test, zero_division=0)
rec = recall_score(test_y, y_pred_test, zero_division=0)

print("Test accuracy :", round(acc, 4))
print("Test precision:", round(prec, 4))
print("Test recall   :", round(rec, 4))
print("\nConfusion matrix (rows=true, cols=pred):")
print(confusion_matrix(test_y, y_pred_test))
print("\nFull classification report:\n")
print(classification_report(test_y, y_pred_test, digits=4))
````

````python
#On the test set the model obtained accuracy X, precision Y and recall Z (replace with actual numbers).
# For credit default detection we usually prioritize recall (to find true defaulters) over precision, because any missing of default results problems in future
````

````python
# view predicted probabilities to understand model confidence
#For DecisionTree, we can call predict_proba to get class probabilities.
#plot two histograms:predicted probability for class 1 when true label is 0 vs when true label is 1.

import matplotlib.pyplot as plt
import numpy as np

probs_test = dt.predict_proba(test_X)[:, 1]   # probability of class 1

plt.figure(figsize=(8,4))
plt.hist(probs_test[test_y==0], bins=25, alpha=0.6, density=True, label='true=0')
plt.hist(probs_test[test_y==1], bins=25, alpha=0.6, density=True, label='true=1')
plt.xlabel('Predicted probability for class 1')
plt.ylabel('Density')
plt.legend()
plt.title('Predicted probability distributions (test)')
plt.show()
````

````python
import pandas as pd
from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Load the Actual Data ---
# This assumes the data fetch command has been executed successfully
cc_defaults = fetch_ucirepo(name='Default of Credit Card Clients')

# Assign Features (X) and Target (Y)
X = cc_defaults.data.features
# The target is the first (and only) column in the 'targets' DataFrame
Y = cc_defaults.data.targets.iloc[:, 0]

# Drop the 'ID' column if it exists, as it is not a feature
if 'ID' in X.columns:
    X = X.drop('ID', axis=1, errors='ignore')

# --- 2. Split the Data ---
# 70% Training, 30% Testing split, stratified for robust class representation
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42, stratify=Y
)

print(" Data Preparation Completed")
print(f"Total samples: {len(X)} | Target Class (1) Proportion: {Y.mean():.4f}")
print(f"Training set size: {len(X_train)} samples")
print(f"Testing set size: {len(X_test)} samples")
print("-" * 50)
````

````python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define the model
dt_tuned = DecisionTreeClassifier(random_state=42)

# Define the hyperparameter grid to search (tuning max_depth and min_samples_leaf)
# We limit max_depth because the default (None) caused severe overfitting.
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_leaf': [10, 20, 50, 100]
}

# Use GridSearchCV to search for the best combination.
# We use 'f1' as the scoring metric since the classes are imbalanced and we need
# a balance between Precision and Recall.
grid_search = GridSearchCV(
    estimator=dt_tuned,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    verbose=1,
    n_jobs=-1
)

# Fit the grid search to the training data
grid_search.fit(train_X, train_y)

# Get the best estimator (the best-performing Decision Tree)
best_dt_model = grid_search.best_estimator_

print("Hyperparameter Tuning Results")
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"Best Cross-Validation F1 Score: {grid_search.best_score_:.4f}")
print("-" * 50)
````
